```python
import pandas as pd
import numpy as np
```

```python
i_cols = ['movie_id', 'movie title' ,'release date','video release date', 'IMDb URL', 'unknown', 'Action', 'Adventure',
    'Animation', 'Children\'s', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy',
    'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']

items = pd.read_csv('ml-100k/u.item', sep = '|', names = i_cols, encoding = 'latin-1')
```

```python
items.tail(10)
```

```python
# Select only needed item
movie_content = items[['movie_id', 'movie title' , 'Action', 'Adventure',
    'Animation', 'Children\'s', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy',
    'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']]
movie_content
```

```python
# Read the ratings dataset
r_cols = ['user_id', 'movie_id', 'rating', 'unix_timestamp']
ratings = pd.read_csv('ml-100k/u.data', sep = '\t', names = r_cols, encoding = 'latin-1')
```

```python
ratings
```

```python
rating_stats = ratings.groupby("movie_id").agg(count=("rating", "count"),avg_rating=("rating", "mean")).reset_index()
rating_stats
```

```python
movies = movie_content.merge(rating_stats, on="movie_id", how="left")
movies
```

# Top Rated Based on Genre

```python
def top_movies_by_genre(genre, metric="avg_rating", n=10):
   
    if genre not in movies.columns:
        print(f"Genre '{genre}' not found in movie data.")
        pass
    else:
    # Filter where genre = 1
        genre_movies = movies[movies[genre] == 1]
        
        # Sort by the chosen metric (descending)
        genre_movies = genre_movies.sort_values(metric, ascending=False)
        
        # Select only important columns
        return genre_movies[["movie_id", "movie title", 'count', metric]].head(n)
```

```python
top_movies_by_genre('Action', metric="avg_rating", n=5)
```

```python
top_movies_by_genre("Adventure", metric="avg_rating", n=5)
```

```python
top_movies_by_genre("Emotional", metric="avg_rating", n=5)
```

```python
top_movies_by_genre("Drama", metric="avg_rating", n=5)
```

# Top Rated Based on Average Rating of whole movie data

```python
def top_movies_by_count(metric="count", n=10):
   
    # Sort by the chosen metric (descending)
    movie = movies.sort_values(metric, ascending=False)
    
    # Select only important columns
    return movie[["movie_id", "movie title", 'count', metric]].head(n)
```

```python
top_movies_by_count(metric="avg_rating", n=5)
```
